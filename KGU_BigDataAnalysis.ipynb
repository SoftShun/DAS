```python
# 이것은 파이썬 한 줄 주석입니다

'''
이것은 파이썬 여러줄 주석 입니다
'''

# math 모듈 포함
import math

# get_area() 함수 정의
def get_area(radius):
  """반지름을 입력 받아 원의 넓이를 반환하는 get_area() 함수"""
  area = math.pi*math.pow(radius,2)
  return area

radius = 1.5
# get_area() 함수 호출 결과를 area 변수에 저장
area = get_area(radius)
print(area)
print(get_area.__doc__) # Docstring 확인
```

```python
import math
import matplotlib.pyplot as plt

x=[]
y_sin=[]
y_cos=[]

for angle in range(360) :
  x.append(angle)
  y_sin.append(math.sin(math.radians(angle)))
  y_cos.append(math.cos(math.radians(angle)))

plt.plot(x,y_sin, label='Sine', color = 'blue')
plt.plot(x,y_cos, label='Cosine', color = 'red')

plt.title('Sine-Cosine Wave')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
# 데이터 스케일링
from sklearn.preprocessing import MinMaxScaler

data = df[['Annual Income (k$)',	'Spending Score (1-100)']]  # 연봉과 소비지수만 출력
data.describe()
```

```python
scaler = MinMaxScaler() # 연봉 편차가 크기 때문에 스케일링
data_sc = scaler.fit_transform(data)
data_df = pd.DataFrame(data_sc)
data_df.describe()
```

```python
from sklearn.cluster import KMeans  # 모델에 훈련시키고 predict에 저장

k = 5

model = KMeans(n_clusters = k, random_state = 10)
model.fit(data_sc)

predict = pd.DataFrame(model.predict(data_sc))
predict.columns=['predict']
```

```python
r = pd.concat([data, predict], axis = 1)
```

```python
r.head()
```

```python
plt.scatter(r['Annual Income (k$)'], r['Spending Score (1-100)'], c = r['predict'], alpha = 0.5)

plt.title('K = %d results ' %k, size = 15)
plt.xlabel('Annual Income', size = 12)
plt.ylabel('Spending Score', size = 12)
plt.grid()

plt.show()
```

```python
da2 = df[['Age', 'Spending Score (1-100)']] # 나이와 소비지수
da2.describe()
```

```python
k = 3

model = KMeans(n_clusters = k, random_state = 10)
model.fit(da2)

predict = pd.DataFrame(model.predict(da2))
predict.columns = ['predict']

r2 = pd.concat([da2, predict], axis = 1)
r2.head()
```

```python
centers = pd.DataFrame(model.cluster_centers_,
                       columns = ['Age', 'Spending Score (1-100)'])
centers_x = centers['Age']
centers_y = centers['Spending Score (1-100)']
```

```python
plt.scatter(r2['Age'], r2['Spending Score (1-100)'], c = r2['predict'], alpha = 0.5)
plt.scatter(centers_x, centers_y, s = 70, marker = 'D', c = '#44a7ff')

plt.title('K = %d results ' %k, size = 15)
plt.xlabel('Age', size = 12)
plt.ylabel('Spending Score (1-100)', size = 12)
```
